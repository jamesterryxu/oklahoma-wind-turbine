```python
# animation line plot example

fig, ax = plt.subplots(1, 1, figsize = (12, 6),dpi=200)

def animate(i):
    ax.cla() # clear the previous image
    ax.plot(luna_position, luna_data.iloc[i][:-1]) # plot the line
    ax.set_xlim([luna_position[0], luna_position[-1]]) # fix the x axis
    ax.set_ylim([-1.1*100, 1.1*100]) # fix the y axis
    ax.text(70, 100,'Time = ' + str(round(luna_data['dt (s)'][i],2)) + ' s')
    ax.set_xlabel('Distance (m)')
    ax.set_ylabel('microStrain')
    

anim = animation.FuncAnimation(fig, animate, frames = len(luna_data.iloc[0][:-1])-2, interval = 1, blit = False)
# plt.show()
# anim.save('orbita.gif', writer=PillowWriter(fps=4))
```

```python
fig, ax = plt.subplots(1, 1, figsize = (12, 6),dpi=200)

def animate_zeroed(i):
    ax.cla() # clear the previous image
    ax.plot(luna_position, luna_data_zeroed.iloc[i][:-1]) # plot the line
    ax.set_xlim([luna_position[0], luna_position[-1]]) # fix the x axis
    ax.set_ylim([-1.1*100, 1.1*100]) # fix the y axis
    ax.text(60, 90,'Time = ' + str(round(luna_data['dt (s)'][i],2)) + ' s')
    ax.set_xlabel('Distance (m)')
    ax.set_ylabel('microStrain')
    ax.axvline(x=35)
    
# frames = len(luna_data.iloc[0][:-1])-2
anim_zeroed = animation.FuncAnimation(fig, animate_zeroed, frames = 1000, interval = 1, blit = False)
# plt.show()
anim_zeroed.save('zeroed_short.gif', writer=PillowWriter(fps=10))
```
